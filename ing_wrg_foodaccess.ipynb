```python
## import dependencies
import urllib2
import os
import openpyxl
import pandas as pd
import csv
```

```python
# Creates directory for data - CBSA - FIPS CROSSWALK
CBSA = os.path.join(os.getcwd(),"CBSA") #folder name - if this doesnt exist then make it
if not os.path.exists(CBSA):
    os.makedirs(CBSA)

filename = CBSA + "/" + "FIPS_CBSA.csv"
url= "http://www.nber.org/cbsa-csa-fips-county-crosswalk/cbsa2fipsxw.csv"

f = urllib2.urlopen(url)
data = f.read()
with open(filename, "wb") as code:
    code.write(data)
```

```python
df = pd.read_csv('FIPS_CBSA.csv', converters={'fipsstatecode': lambda x: str(x) ,
                                              'fipscountycode': lambda x: str(x) } )
```

```python
df [['fipsstatecode', "cbsacode", 'fipscountycode']] #view obs
```

```python
df["FIPS"] = df["fipsstatecode"].map(str) + df["fipscountycode"]
```

```python
CBSA = df[['FIPS','cbsacode']]
```

```python
CBSA.head(5)
```

```python
#REMOVE DUPLICATES FROM CBSA FILE
CBSA.drop_duplicates(['cbsacode'],)

#REMOVE MISSING VALUES
CBSA[pd.notnull(CBSA['cbsacode'])] 
```

```python
FINAL_FOOD = pd.merge(FOOD, CBSA, on='FIPS')
```

```python
FINAL_FOOD
```

```python
#REMOVE DUPLICATES FROM CBSA FILE
final = FINAL_FOOD.drop_duplicates(['cbsacode'],)
```

```python
food_access = final[['PCT_LACCESS_POP10', 'PCT_LACCESS_LOWI10', 'PCT_LACCESS_CHILD10', 'PCT_LACCESS_SENIORS10', 'PCT_LACCESS_HHNV10', 'cbsacode']]
```
